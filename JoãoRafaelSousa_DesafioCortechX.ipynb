# **<font face="Times New Roman">Questão 1: Diferença Máxima em Registros de Manutenção de Frota</font>**



## <font face="Times New Roman"> Descrição do Cenário </font>

---


<font face="Times New Roman" size="4" > A CortechX opera uma frota corporativa composta exclusivamente por sedans de médio porte,
distribuídos entre as filiais da empresa. Os modelos gerenciados atualmente são:
<b>Honda Civic</b>, <b>Honda City</b>, <b>Nissan Versa</b> e <b>Mitsubishi Lancer</b>.
</font>

<font face="Times New Roman" size="4" >Cada veículo da frota passa por revisões periódicas, e o sistema de gestão registra,
para cada evento de manutenção, o custo total da intervenção em reais (R$), armazenado
como um valor inteiro positivo. Esses registros são consolidados diariamente em um único
array linear, na ordem cronológica em que as ordens de serviço foram abertas.

<font face="Times New Roman" size="4" >A equipe de controladoria identificou a necessidade de um novo módulo na API interna de
análise financeira. Esse módulo deve ser capaz de, dado o array de custos de manutenção
do dia, determinar o **maior lucro potencial de arbitragem interna** entre dois registros:
ou seja, encontrar a diferença máxima entre um valor que ocorre **antes** de outro no
array, de forma que o valor posterior seja maior que o anterior.

<font face="Times New Roman" size="4" >Formalmente, dado um array `custos[]` de tamanho `n`, o sistema deve retornar o valor
máximo de `custos[j] - custos[i]`, onde `j > i` e `custos[j] > custos[i]`.
Caso nenhum par válido exista, o sistema deve retornar `-1`.

</font>

````python
def calculate_max_diff(custos):
    '''Para que o código rode bem com até 100 mil registros, fiz com que minha função percorresse o array apenas 1 vez (apenas 1 for),
    mantendo o controle do menor valor e da maior diferença em toda a iteração, evitando fazer 2 loops (O(n²)).'''
    
    # Caso possua 1 elemento, retorna -1
    if len(custos) < 2:
        return -1
    
    # Inicializando o valor mínimo com o primeiro elemento e a diferença com -1
    min_custo = custos[0]
    max_diff = -1
    
    # Percorrendo array partir do 2º elemento
    for j in range(1, len(custos)):
        custo_atual = custos[j]

        # Se o custo atual for maior que o mínimo já visto, calculamos a diferença
        if custo_atual > min_custo: 
            diff = custo_atual - min_custo
            if diff > max_diff: # Se a diferença atual for a maior que já vimos
                max_diff = diff # A diferença máxima atualiza
        
        if custo_atual < min_custo: # Se o custo atual for o menor que já vimos...
            min_custo = custo_atual # Atualizamos, independente de ter calculado a diferença
    
    return max_diff
````

````python
# Teste da Função
exemplo = [340, 120, 490, 85, 600, 210]
result = calculate_max_diff(exemplo)
print(f"Maior diferença encontrada: {result}")
````

### Decisões da Questão 1

**Estratégia adotada:** Para garantir o processamento eficiente de até 100.000 registros dentro do rigoroso limite de tempo (200 ms), a solução foi desenhada para iterar sobre o array apenas uma vez.

**Por que isso é melhor?** 
Evitamos a abordagem força-bruta tradicional de laços duplos aninhados, reduzindo a **Complexidade de Tempo** de `O(n²)` para um ótimo `O(n)`. Paralelamente, utilizamos apenas duas variáveis de controle (`min_custo` e `max_diff`), garantindo uma **Complexidade de Espaço** constante de `O(1)`.

  * <font face="Times New Roman" size="4">**OuroImperial (Camada Gold):** Agrega os dados limpos pelo `transporte_id` e retorna um dicionário com: *total de missões*, *receita total de dracmas* e a *velocidade média* (km/s = distancia_km / duracao_viagem_s).</font>

#### <font face="Times New Roman">2.2.1 Imports e Strategy Pattern (A Fonte das Mensagens)</font>

<font face="Times New Roman" size="4">Nesta célula, importamos as bibliotecas necessárias e definimos como vamos ler as mensagens que chegam ao Acampamento.</font>

````python
import pandas as pd
from abc import ABC, abstractmethod
from typing import List, Dict, Any, Optional

class FonteDados(ABC):
    """Classe abstrata para definição da estratégia de leitura de dados."""
    @abstractmethod
    def ler(self) -> List[Dict[str, Any]]:
        pass

class FontePergaminho(FonteDados):
    """Estratégia para ler dados de uma lista em memória (cache mágico)."""
    def __init__(self, dados: List[Dict[str, Any]]):
        self._dados = dados

    def ler(self) -> List[Dict[str, Any]]:
        return self._dados

class FonteCSV(FonteDados):
    """Estratégia para ler dados de um arquivo CSV (Portal de NYC)."""
    def __init__(self, path: str, nrows: Optional[int] = None):
        self._path = path
        self._nrows = nrows

    def ler(self) -> List[Dict[str, Any]]:
        # Lendo o CSV e convertendo para lista de dicionários
        df = pd.read_csv(self._path, nrows=self._nrows)
        return df.to_dict('records')
````

#### <font face="Times New Roman">2.2.2 Camadas Medallion (Os Metais Divinos)

<font face="Times New Roman" size="4">Aqui, forjamos as três camadas de processamento. Cada camada é responsável por um nível de refinamento dos dados.</font>

````python
class Camada(ABC):
    """Classe abstrata para as camadas da arquitetura Medallion."""
    @abstractmethod
    def processar(self, dados: List[Dict[str, Any]]) -> Any:
        pass

class BronzeCelestial(Camada):
    """Camada Bronze: Extração de dados brutos da mina."""
    def processar(self, dados: List[Dict[str, Any]]) -> List[Dict[str, Any]]:
        return dados

class PrataEstigia(Camada):
    """Camada Silver: Limpeza e padronização dos dados (Purificação)."""
    def processar(self, dados: List[Dict[str, Any]]) -> List[Dict[str, Any]]:
        dados_limpos = []
        for registro in dados:
            # Mapeamento de colunas (CSV -> Scroll/Mágico)
            dracmas = registro.get('dracmas_pagos') if 'dracmas_pagos' in registro else registro.get('fare_amount')
            distancia = registro.get('distancia_km') if 'distancia_km' in registro else registro.get('trip_distance')
            passageiros = registro.get('passageiros_semideuses') if 'passageiros_semideuses' in registro else registro.get('passenger_count')
            
            # Regras de Negócio: Filtragem de registros corrompidos
            if dracmas is None or distancia is None or passageiros is None:
                continue
            if distancia <= 0 or passageiros == 0:
                continue
            
            # Mantendo nomes padronizados para a camada Gold
            registro_limpo = registro.copy()
            registro_limpo['dracmas_pagos'] = dracmas
            registro_limpo['distancia_km'] = distancia
            registro_limpo['passageiros_semideuses'] = passageiros
            # Padronizando ID de transporte
            registro_limpo['transporte_id'] = registro.get('transporte_id') or registro.get('vendor_id')
            
            dados_limpos.append(registro_limpo)
        return dados_limpos

class OuroImperial(Camada):
    """Camada Gold: Agregação e indicadores de negócio (Relatório Consolidado)."""
    def processar(self, dados: List[Dict[str, Any]]) -> Dict[str, Dict[str, Any]]:
        # Agrupamento por transporte_id
        agregado = {}
        for r in dados:
            tid = str(r.get('transporte_id', 'Desconhecido'))
            duracao = r.get('duracao_viagem_s') or r.get('trip_duration', 0)
            
            if tid not in agregado:
                agregado[tid] = {'missoes': 0, 'receita': 0.0, 'distancia_total': 0.0, 'duracao_total': 0}
            
            agregado[tid]['missoes'] += 1
            agregado[tid]['receita'] += float(r['dracmas_pagos'])
            agregado[tid]['distancia_total'] += float(r['distancia_km'])
            agregado[tid]['duracao_total'] += int(duracao)

        # Calculando indicadores finais
        resultado = {}
        for tid, stats in agregado.items():
            vel_media = 0.0
            if stats['duracao_total'] > 0:
                vel_media = stats['distancia_total'] / stats['duracao_total']
            
            resultado[tid] = {
                'total_missões': stats['missoes'],
                'receita_total': round(stats['receita'], 2),
                'velocidade_media': round(vel_media, 6)
            }
        return resultado
````

#### <font face="Times New Roman">2.2.3 Factory Pattern (A Forja)</font>
<font face="Times New Roman" size="4">Esta célula contém a fábrica estática que vai instanciar a camada correta dependendo do metal divino solicitado pelo motor.</font>

````python
class ForjaCamadaFactory:
    """Fábrica para instanciar as camadas da Forja de Hefesto."""
    @staticmethod
    def criar(tipo: str) -> Camada:
        if tipo == "bronze_celestial":
            return BronzeCelestial()
        elif tipo == "prata_estigia":
            return PrataEstigia()
        elif tipo == "ouro_imperial":
            return OuroImperial()
        else:
            raise ValueError(f"Metal desconhecido: {tipo}. A Forja não reconhece esse material.")
````

#### <font face="Times New Roman">2.2.4 Motor de Ingestão de Hefesto</font>

<font face="Times New Roman" size="4">Esta é a engrenagem principal. O motor vai conectar a fonte de dados com a forja de camadas.</font>

````python
class MotorHefesto:
    """Motor de Ingestão: Orquestra o fluxo de dados entre as camadas."""
    def __init__(self, fonte: FonteDados):
        self.fonte = fonte
        self.factory = ForjaCamadaFactory()

    def executar(self):
        print("\n--- [FORJA DE HEFESTO] Iniciando Ingestão Mágica ---")
        
        # 1. Leitura
        dados_brutos = self.fonte.ler()
        
        # 2. Bronze
        bronze = self.factory.criar("bronze_celestial")
        dados_bronze = bronze.processar(dados_brutos)
        print(f"[BRONZE CELESTIAL] Registros extraídos: {len(dados_bronze)}")
        
        # 3. Prata
        prata = self.factory.criar("prata_estigia")
        dados_prata = prata.processar(dados_bronze)
        print(f"[PRATA ESTÍGIA] Registros purificados: {len(dados_prata)}")
        
        # 4. Ouro
        ouro = self.factory.criar("ouro_imperial")
        dados_ouro = ouro.processar(dados_prata)
        
        print("\n[OURO IMPERIAL] Relatório Consolidado de Missões:")
        for tid, info in dados_ouro.items():
            print(f"Transporte: {tid:<15} | Missões: {info['total_missões']:<5} | "
                  f"Receita: {info['receita_total']:>10.2f} dracmas | "
                  f"Vel. Média: {info['velocidade_media']:.6f} km/s")
        
        print("--- [FORJA DE HEFESTO] Processamento Concluído ---")
        return dados_ouro
````

#### <font face="Times New Roman">2.2.5 Execução </font>

<font face="Times New Roman" size="4">Por fim, a célula de execução. Aqui estão os dados brutos e a inicialização do motor para testar se a sua forja está funcionando corretamente.</font>

````python
# Inicializando a fonte de dados com o CSV de Nova York
fonte = FonteCSV('data/NYC.csv', nrows=5000)

# Inicializando o Motor de Ingestão de Hefesto
motor = MotorHefesto(fonte)

# Executando o pipeline completo
resultado_final = motor.executar()
````

###  Decisões da Questão 2

**Design Patterns aplicados:** O pipeline foi desenhado utilizando Orientação a Objetos para garantir que o "Motor de Hefesto" seja totalmente extensível:

* **Strategy:** A classe abstrata `FonteDados` permite plugar novas fontes de ingestão (como memória local ou CSV) no futuro sem alterar a lógica do motor principal.
* **Factory:** A classe `ForjaCamadaFactory` centraliza a criação dos objetos das camadas Medallion, garantindo que o orquestrador instancie a camada correta apenas passando uma string.

> **Resumo:** Essa separação de responsabilidades (SRP) isola a regra de negócio da extração de dados, deixando o código limpo, resiliente e totalmente preparado para escala.

### <font face="Times New Roman"> 3.5.1 Preparação do dataset</font>

<font face="Times New Roman" size="4">Antes de iniciar os modelos, é necessário preparar o dataset para uso em Machine Learning.</font>

<font face="Times New Roman" size="4">Considere os tratamentos realizados na Questão 3 e adapte o conjunto de dados conforme o problema que será desenvolvido. Nesta etapa, os dados devem ser organizados de forma que possam ser utilizados por algoritmos de aprendizado, o que pode envolver ajustes nas variáveis, separação entre entradas e saída, e definição de conjuntos de treino e teste.</font>

<font face="Times New Roman" size="4">Dependendo da escolha dos targets, pode ser necessário preparar versões diferentes do dataset para cada problema.</font>


````python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Drop de possíveis NaNs residuais gerados no feature engineering para não quebrar o fit do sklearn
df_model = df_ml.dropna().copy()

# Se sobrou alguma coluna de texto (object), converte para ID numérico
colunas_texto = df_model.select_dtypes(include=['object']).columns
for col in colunas_texto:
    df_model[col] = df_model[col].astype('category').cat.codes

# 1. Setup de Classificação (Target: is_legendary)
y_clf = df_model['is_legendary']

# Drop de features altamente correlacionadas ('capture_rate', 'is_mythical') para evitar Data Leakage
X_clf = df_model.drop(columns=['is_legendary', 'is_mythical', 'capture_rate'], errors='ignore')

# Divisão 80/20 com estratificação devido ao forte desbalanceamento de classes do target
X_clf_train, X_clf_test, y_clf_train, y_clf_test = train_test_split(
    X_clf, y_clf, test_size=0.2, random_state=42, stratify=y_clf
)


# 2. Setup de Regressão (Target: capture_rate)
y_reg = df_model['capture_rate']
X_reg = df_model.drop(columns=['capture_rate'], errors='ignore')

# Divisão simples 80/20 para regressão contínua
X_reg_train, X_reg_test, y_reg_train, y_reg_test = train_test_split(
    X_reg, y_reg, test_size=0.2, random_state=42
)


# 3. Normalização
# Fit/Transform no treino, apenas Transform no teste para não vazar a média real do teste
scaler_clf = StandardScaler()
X_clf_train_scaled = scaler_clf.fit_transform(X_clf_train)
X_clf_test_scaled = scaler_clf.transform(X_clf_test)

scaler_reg = StandardScaler()
X_reg_train_scaled = scaler_reg.fit_transform(X_reg_train)
X_reg_test_scaled = scaler_reg.transform(X_reg_test)

print("Dados preparados e normalizados!")
print(f"Features de Classificação: {X_clf_train_scaled.shape[1]}")
````

### **Modelo 1: Classificação**

**Objetivo:** 
* Prever se um Pokémon é classificado como Lendário (`is_legendary`) utilizando exclusivamente seus atributos físicos, estatísticas de combate e as variáveis matemáticas geradas (Feature Engineering).

**Formulação Técnica:**
* Como a grande maioria dos Pokémon não é lendária, caímos em um problema clássico de dados desbalanceados. Para lidar com isso sem precisar criar dados falsos, decidi usar o algoritmo Random Forest com o parâmetro `class_weight='balanced'`. Dessa forma, o próprio modelo entende que os lendários são raros e dá um peso maior para eles durante o treinamento.

**A abordagem adotada consiste em:**
* **Prevenção de Data Leakage:** Exclusão de variáveis diretamente correlacionadas à raridade (como `capture_rate` e `is_mythical`) para garantir que o modelo aprenda pelos padrões estruturais.
* **Algoritmo Robusto:** Utilização do `RandomForestClassifier`. A escolha se deve ao excelente desempenho do algoritmo com dados tabulares e à possibilidade de aplicar o hiperparâmetro `class_weight='balanced'`, que penaliza o erro na classe minoritária e dispensa o uso de geradores de dados sintéticos (como o SMOTE).

````python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# Instancia o modelo com balanceamento interno para a classe minoritária
clf_model = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)

# Treinamento do classificador
clf_model.fit(X_clf_train_scaled, y_clf_train)

# Previsão utilizando os dados de teste
y_clf_pred = clf_model.predict(X_clf_test_scaled)

# Avaliação das métricas do modelo
print("--- RELATÓRIO DE CLASSIFICAÇÃO ---")
print(classification_report(y_clf_test, y_clf_pred, target_names=['Comum', 'Lendário']))
````

<font face="Times New Roman" size="4">Escolha uma variável do dataset que possa ser interpretada como um valor a ser estimado. A escolha deve fazer sentido dentro do contexto dos dados e apresentar variabilidade suficiente para justificar o uso de regressão.</font>

### **Modelo 2: Regressão**

**Objetivo:** 
* Prever a Taxa de Captura (`capture_rate`) de um Pokémon utilizando seus atributos de combate, resiliência, peso e as variáveis matemáticas geradas (Feature Engineering).

**Formulação Técnica:**
* Diferente da classificação, aqui temos um problema contínuo, onde o modelo precisa prever um número exato (que varia de 3, para Pokémon muito difíceis, até 255, para os mais fáceis). O grande desafio técnico é que a relação entre a força de um Pokémon e a dificuldade de capturá-lo não forma uma reta perfeita, sofrendo quebras muito bruscas que fogem do padrão matemático simples.

**A abordagem adotada consiste em:**
* **Prevenção de Linearidade Simples:** Decidi não usar uma Regressão Linear tradicional. Como as lendas do jogo, por exemplo, têm uma queda abrupta na taxa de captura, um modelo que tenta traçar uma linha reta não conseguiria acompanhar essas variações irregulares com precisão.
* **Algoritmo Robusto:** Utilização do `RandomForestRegressor`. A escolha se deve ao fato de que modelos baseados em árvore lidam perfeitamente com essas quebras não-lineares. Eles conseguem cruzar informações complexas (como o Bulk Físico e os status totais) de uma forma bem mais inteligente para chegar na taxa exata.


````python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# --- Instanciação do Modelo ---
# O Random Forest Regressor cria várias árvores e tira a média numérica das previsões delas
reg_model = RandomForestRegressor(n_estimators=100, random_state=42)

# --- Treinamento ---
# O modelo mapeia os números contínuos do target no conjunto de treino
reg_model.fit(X_reg_train_scaled, y_reg_train)

# --- Predição ---
# Prevendo a taxa de captura na base isolada de teste
y_reg_pred = reg_model.predict(X_reg_test_scaled)

# --- Resultado Rápido ---
r2 = r2_score(y_reg_test, y_reg_pred)
mae = mean_absolute_error(y_reg_test, y_reg_pred)
rmse = np.sqrt(mean_squared_error(y_reg_test, y_reg_pred))

print("--- BOLETIM DO MODELO DE REGRESSÃO ---")
print(f"R² (Poder de Explicação): {r2:.4f}")
print(f"MAE (Erro Médio Absoluto): {mae:.2f} pontos de taxa")
print(f"RMSE (Erro Quadrático Médio): {rmse:.2f} pontos de taxa")
````

````python
COTACAO_FIXA = {
    "USD": 5.18,  # Dólar Americano para Real
    "EUR": 5.54,  # Euro para Real
    "BRL": 1.00   # Moeda base
}
````

````python
import requests
import time

# Configurações iniciais
ENDPOINT_INVENTARIO = "https://api-4ifx.onrender.com/v2/inventory"
HEADERS = {"X-API-KEY": "teste-2024"}

def gerar_relatorio_guilda():
    cursor = None
    all_processed = False
    
    print("--- Iniciando a Quest no Inventário (Aguardando o Portal despertar...) ---")
    
    while not all_processed:
        params = {}
        if cursor:
            params["cursor"] = cursor
            
        try:
            response = requests.get(ENDPOINT_INVENTARIO, headers=HEADERS, params=params)
            
            # --- TRATAMENTO DO RATE LIMIT (Erro 429) ---
            if response.status_code == 429:
                print("! Feitiço de Rate Limit detectado. Recuando por 3 segundos...")
                time.sleep(3)
                continue
                
            if response.status_code != 200:
                print(f"Erro inesperado no portal: {response.status_code}. Tentando reconectar...")
                time.sleep(2)
                continue

            data = response.json()
            itens = data.get("items", [])
            
            # --- PROCESSAMENTO DOS ITENS ---
            for item in itens:
                nome = item.get("item")
                preco_raw = item.get("preco")
                moeda = item.get("moeda", "BRL")
                categoria = item.get("categoria", "Outros")
                
                if preco_raw is None:
                    continue
                
                try:
                    # Trata vírgulas e converte para float
                    preco_limpo = str(preco_raw).replace(",", ".")
                    preco_float = float(preco_limpo)
                    
                    # Conversão para BRL
                    valor_brl = preco_float * COTACAO_FIXA.get(moeda, 1.0)
                    
                    print(f"Produto: {nome:<30} | Categoria: {categoria:<15} | Valor: R$ {valor_brl:.2f}")
                except ValueError:
                    continue

            # --- LÓGICA DE PAGINAÇÃO ---
            cursor = data.get("next_page_token")
            if not cursor:
                all_processed = True
                print("\n--- Missão Cumprida: Todo o inventário foi processado! ---")

        except Exception as e:
            print(f"Falha na conexão com a Dungeon: {e}. Tentando novamente em 5 segundos...")
            time.sleep(5)

gerar_relatorio_guilda()
````

### Resumo da Questão 4

**Estratégias de Integração adotadas:** O pipeline de extração foi desenhado para ser robusto e autossuficiente contra instabilidades e limitações da API (a "Dungeon"):

* **Tratamento de Rate Limit (Erro 429):** Foi implementado um mecanismo de defesa que intercepta o status `429 (Too Many Requests)`. O sistema compreende o bloqueio temporário, pausa a execução (`time.sleep`) e retoma a extração automaticamente, garantindo que o programa não "quebre" de forma abrupta.
* **Paginação com Cursor (Tokens):** O laço de repetição foi atrelado ao `next_page_token`, permitindo varrer todas as páginas da API de forma sequencial até que o inventário inteiro seja coletado.
* **Data Cleaning & Normalização "On the Fly":** Durante a extração, o script já realiza a higienização de dados inconsistentes (como vírgulas em vez de pontos flutuantes) e aplica a conversão de moedas estrangeiras para Reais (BRL) usando um dicionário central de câmbio de forma totalmente dinâmica.

### <font face="Times New Roman" size="5">Tarefa 4.1 - Frequência de Visitas ao Apartamento 1201</font>

Liste todos os visitantes que estiveram no apartamento **1201** nos últimos 6 meses, informando quantas vezes cada um visitou, a data da primeira visita, a data da última visita e os motivos registrados. Ordene pelo número de visitas de forma decrescente.

````python
query = text("""
SELECT 
    v.visitante_nome,
    COUNT(v.id) AS total_visitas,             -- Conta quantas vezes visitou
    MIN(v.data_entrada) AS primeira_visita,  -- Data da primeira vez
    MAX(v.data_entrada) AS ultima_visita,    -- Data da última vez
    STRING_AGG(DISTINCT v.motivo, ' | ') AS motivos_registrados -- Reúne motivos sem repetir
FROM visitas v
JOIN apartamentos a ON v.apartamento_id = a.id
WHERE a.numero = '1201' 
  -- Filtra 6 meses antes da última data registrada no banco
  AND v.data_entrada >= (SELECT MAX(data_entrada) FROM visitas) - INTERVAL '6 months'
GROUP BY v.visitante_nome
ORDER BY total_visitas DESC;
""")

df_frequencia_visitas = pd.read_sql(query, engine)
df_frequencia_visitas
````

### <font face="Times New Roman" size="5">Tarefa 4.4 - CTE: Linha do Tempo Completa do Suspeito</font>

<font face="Times New Roman" size="4">

Usando uma CTE (Common Table Expression), construa a linha do tempo completa de **Rodrigo Sampaio** no dia 15/11/2024, reunindo em uma única consulta seus registros de câmera, suas visitas e as ocorrências que ele gerou naquele dia. Ordene cronologicamente.

</font>

````python
query = text("""
WITH AtividadeSuspeito AS (
    -- 1. Registros de Câmera (Morador)
    SELECT 
        rc.timestamp AS horario, 
        rc.descricao, 
        'Câmera' AS tipo_registro
    FROM registros_camera rc
    JOIN moradores m ON rc.morador_id = m.id
    WHERE m.nome = 'Rodrigo Sampaio' AND rc.timestamp::date = '2024-11-15'
    
    UNION ALL
    
    -- 2. Livro de Visitas (Ele como síndico visitando o 1201)
    SELECT 
        v.data_entrada AS horario, 
        'Visita ao apt ' || a.numero || ': ' || v.motivo AS descricao, 
        'Visita' AS tipo_registro
    FROM visitas v
    JOIN apartamentos a ON v.apartamento_id = a.id
    WHERE v.visitante_nome = 'Rodrigo Sampaio' AND v.data_entrada::date = '2024-11-15'
    
    UNION ALL
    
    -- 3. Ocorrências (Registradas por ele ou onde ele é o foco)
    SELECT 
        o.data_hora AS horario, 
        o.descricao, 
        'Ocorrência' AS tipo_registro
    FROM ocorrencias o
    LEFT JOIN funcionarios f ON o.registrado_por = f.id
    WHERE (f.nome = 'Rodrigo Sampaio' OR o.descricao ILIKE '%Rodrigo Sampaio%') 
      AND o.data_hora::date = '2024-11-15'
)
SELECT * FROM AtividadeSuspeito
ORDER BY horario;
""")

df_timeline = pd.read_sql(query, engine)
df_timeline
````

### <font face="Times New Roman" size="5">Tarefa 5.1 - Window Function: Frequência de Visitas ao Apt 1201 por Visitante</font>

<font face="Times New Roman" size="4"> Para cada visita registrada ao apartamento **1201**, exiba o nome do visitante, a data da visita e o número acumulado de visitas daquele visitante até aquela data, usando uma window function. Ordene por visitante e depois por data.

</font>

````python
query = text("""
SELECT 
    v.visitante_nome, 
    v.data_entrada,
    -- Window Function: Conta as visitas acumuladas por visitante ao longo do tempo
    COUNT(*) OVER (
        PARTITION BY v.visitante_nome 
        ORDER BY v.data_entrada
    ) AS visita_numero_acumulado
FROM visitas v
JOIN apartamentos a ON v.apartamento_id = a.id
WHERE a.numero = '1201' -- Foca no apartamento da vítima
ORDER BY v.visitante_nome, v.data_entrada;
""")

df_frequencia_acumulada = pd.read_sql(query, engine)
df_frequencia_acumulada
````

````python
import cv2
import time
from utils.gesture_utils import HandDetector, ScreenMapper, GestureController

# 1. Inicializando os componentes especializados do pipeline
detector = HandDetector()
mapper = ScreenMapper(suavizacao=0.45)
controller = GestureController()

cap = cv2.VideoCapture(0)
print("Motor de Gestos CortechX iniciado. Pressione ESC para sair.")

while cap.isOpened():
    success, frame = cap.read()
    if not success: 
        break

    # Inverte o frame horizontalmente (efeito espelho)
    frame = cv2.flip(frame, 1)
    h, w = frame.shape[:2]

    # --- PIPELINE DE PROCESSAMENTO ---

    # [FASE 1: MAPEAR] Detecta landmarks via MediaPipe utilizando o utilitário
    lms, side = detector.get_landmarks(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))

    if lms:
        # Identificação matemática dos dedos levantados [polegar, indicador, médio, anelar, mindinho]
        # Polegar exige lógica de lateralidade (X) baseada na mão detectada; os demais usam altura relativa (Y)
        p = lms
        pol = p[4].x < p[3].x if side == "Left" else p[4].x > p[3].x
        fingers = [
            pol,             # Polegar
            p[8].y < p[6].y,   # Indicador
            p[12].y < p[10].y, # Médio
            p[16].y < p[14].y, # Anelar
            p[20].y < p[18].y  # Mindinho
        ]

        # [FASE 2: TRANSLADAR] Converte ponto do indicador para coordenadas da tela com suavização
        cx, cy = mapper.map_to_screen(p[8], w, h)

        # [FASE 3: INTERAGIR] Executa a ação do sistema baseada no gesto detectado
        controller.execute(fingers, (cx, cy), p[8].y * h)

        # Feedback Visual: Desenha a ponta do indicador no frame
        cv2.circle(frame, (int(p[8].x * w), int(p[8].y * h)), 8, (255, 0, 255), -1)

    # Exibe o feedback visual em tempo real
    cv2.imshow("Motor de Gestos - CortechX", frame)
    
    # Sai do loop ao pressionar a tecla ESC
    if cv2.waitKey(1) & 0xFF == 27: 
        break

# Limpeza de recursos
cap.release()
cv2.destroyAllWindows()
````

````python
import os
import json
import logging
from dotenv import load_dotenv
from abc import ABC, abstractmethod
from typing import Any, Dict, List, Optional
import google.generativeai as genai
from google.generativeai.types import GenerationConfig

# Configuração de Log para monitorar o que os agentes estão fazendo
# Adicionamos force=True para garantir que o log apareça corretamente no Colab
logging.basicConfig(level=logging.INFO, format='%(levelname)s: %(message)s', force=True)

# --- ABSTRAÇÃO DE INFRAESTRUTURA (DIP) ---      
class LLMProvider(ABC):       
    @abstractmethod
    def generate_json(self, prompt: str) -> Dict[str, Any]:
        pass

class GeminiProvider(LLMProvider):  
    """Implementação específica para o Google Gemini com JSON forçado."""   
    def __init__(self, model_name: str = "gemini-2.5-flash"):
        self.model = genai.GenerativeModel(        
            model_name=model_name,        
            generation_config=GenerationConfig(
                temperature=0.0, # Determinismo total
                response_mime_type="application/json"
            )
        )

    def generate_json(self, prompt: str) -> Dict[str, Any]:
        try:
            response = self.model.generate_content(prompt)
            return json.loads(response.text)     
        except Exception as e:
            logging.error(f"Erro na API do Gemini: {e}")
            raise

# --- ABSTRAÇÃO DOS AGENTES (SRP - Single Responsibility Principle) ---
class Agent(ABC):
    def __init__(self, llm: LLMProvider):
        self.llm = llm        

    @abstractmethod
    def execute(self, input_data: Any) -> Any:      
        pass

class ExtractorAgent(Agent):  
    """Passo 1: Converte texto bruto em dados estruturados."""
    def execute(self, email_text: str) -> Dict[str, Any]:
        prompt = f"""
        Extraia as entidades do e-mail abaixo no formato JSON estrito:
        - "titulo_projeto": (String)
        - "orientador": (String)
        - "area_foco": (String)
        - "tecnologias_requeridas": (Lista de Strings)

        E-mail: {email_text}  
        """
        return self.llm.generate_json(prompt)

class EvaluatorAgent(Agent):  
    """Passo 2: Aplica as regras de negócio da CortechX."""
    def execute(self, project_data: Dict[str, Any]) -> Dict[str, Any]:
        prompt = f"""
        Avalie o projeto abaixo baseado na Regra de Negócio da CortechX:
        Regra: Aprovado se tecnologia_requeridas tiver linguagens (Python, C++, SQL, etc) ou infra (Docker, Airflow).
        Reprovado se for administrativo ou sem stack técnico.

        Dados: {json.dumps(project_data, ensure_ascii=False)}    

        Retorne JSON: {{"aprovado": bool, "justificativa": str}}        
        """
        return self.llm.generate_json(prompt)

# --- ORQUESTRADOR (Workflow Management) ---
class ProjectTriagePipeline:  
    """Orquestra o fluxo de dados entre os agentes de forma sequencial."""
    def __init__(self, llm: LLMProvider):
        self.extractor = ExtractorAgent(llm)
        self.evaluator = EvaluatorAgent(llm)

    def process(self, email_text: str) -> Dict[str, Any]:
        logging.info("Iniciando extração de dados...")        
        dados = self.extractor.execute(email_text)

        logging.info("Iniciando avaliação técnica...")        
        decisao = self.evaluator.execute(dados) 

        return {
            "metadata": dados,
            "decisao_final": decisao
        }
````

### Resumo da Questão 7

**Estratégia e Design Patterns:** O desafio de triagem de e-mails foi solucionado através de um sistema multiagente, construído sobre princípios de Engenharia de Software:

* **Inversão de Dependência (DIP):** Criou-se a interface abstrata `LLMProvider`. O sistema de agentes não tem acoplamento forte com o provedor de IA. Isso permite trocar de modelo (Groq, OpenAI, etc) no futuro apenas criando uma nova classe, sem alterar a regra de negócio.
* **Responsabilidade Única (SRP):** Os agentes foram isolados em classes especialistas. O `ExtractorAgent` atua exclusivamente na formatação de texto bruto para JSON. Em seguida, o `EvaluatorAgent` consome esses dados estruturados apenas para validar a aprovação dos projetos.
* **Pipeline de Orquestração:** A classe `ProjectTriagePipeline` gerencia o fluxo, garantindo que a saída do Passo 1 seja a entrada do Passo 2 de forma automática, mantendo o ambiente de execução totalmente limpo.

> **Obs:** Para a execução deste pipeline, foi instanciada e utilizada a **API do Google Gemini** (através do modelo `gemini-2.5-flash`). A chave foi configurada de forma segura via Secrets do Colab / Variáveis de Ambiente, garantindo as melhores práticas de cibersegurança.
